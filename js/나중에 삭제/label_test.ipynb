```python
from PIL import Image
import os
import natsort
import csv
import re

img_size = (32, 32)

def resize_img(img_path):
    img_lists = os.listdir(img_path)
    img_lists = natsort.natsorted(img_lists)
    for img_name in img_lists:
        print(f'{img_path}{img_name}')
        image = Image.open(f'{img_path}{img_name}')
        image = image.resize(img_size)
        image.save(f'{img_path}{img_name}')

resize_img('/content/drive/MyDrive/image_recognition/js/train')
resize_img('/content/drive/MyDrive/image_recognition/js/valid')
resize_img('/content/drive/MyDrive/image_recognition/js/test')
```
