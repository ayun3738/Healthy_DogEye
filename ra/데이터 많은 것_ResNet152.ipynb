```python
from tensorflow.keras.applications.resnet import ResNet34


## 일부 재학습 모델
base_model34 = ResNet34(include_top=False, input_shape = (224,224 ,3), weights = 'imagenet')

# base_model.trainable = False

# for layer in base_model.layers[-10:]: 
#     layer.trainable = True				

base_model34.trainable = True

for layer in base_model34.layers[:-10]:
    layer.trainable = False			

for layer in base_model34.layers[:]:	
  print(layer.name, layer.trainable)

inputs = tf.keras.Input(shape=(224, 224, 3))

x = base_model34(inputs, training=False) # batchnorm 부분 update 방지

x = tf.keras.layers.Flatten(input_shape=base_model34.output_shape[1:])(x)
x = tf.keras.layers.Dense(256, activation='relu')(x)
x= tf.keras.layers.Dropout(0.5)(x)
outputs = tf.keras.layers.Dense(11, activation='softmax')(x)

model_res50 = tf.keras.Model(inputs, outputs)
```
